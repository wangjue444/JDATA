```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
comments=pd.read_csv("data_ori/JData_Comment.csv",header=0,iterator=False)
```

```python
comments.head()
```

```python
udt=comments['dt'].unique()
```

```python
grouped=comments.groupby(['sku_id'],as_index=False)
```

```python
def maxandmin(group):
    re=group
    re['max_comment_num']=max(group['comment_num'])
    re['min_comment_num']=min(group['comment_num'])
    re['max_bad_num']=max(group['has_bad_comment'])
    re['min_bad_num']=min(group['has_bad_comment'])
    re['max_badrate']=max(group['bad_comment_rate'])
    re['min_badrate']=min(group['bad_comment_rate'])

    return re.iloc[0][['sku_id','max_comment_num','min_comment_num','max_bad_num','min_bad_num','max_badrate','min_badrate']]
    
```

```python
transcomment=grouped.apply(maxandmin)
```

```python
tmp=transcomment[transcomment['max_comment_num']-transcomment['min_comment_num']!=0]['sku_id']
```

```python
product=pd.read_csv("data_ori/JData_Product.csv",header=0)
```

```python
cou=0
for i in tmp:
    if i in product['sku_id']:
        cou=cou+1
        
```
